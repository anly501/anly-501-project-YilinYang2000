For the infringed, due to the particularity of Internet infringement, the corresponding rights protection process is increasingly difficult.

2. 3. Due to the openness of the Internet, the occurrence of an infringement is likely to cross national borders unconsciously by the infringer, involving multiple regions and countries. At present, when there is no international uniform standard for privacy infringement, it is obviously more difficult for different countries and regions to deal with the same privacy infringement without conflict. Therefore, it can be seen that there is another difficulty in safeguarding the rights and interests of infringing objects in the Internet environment.
